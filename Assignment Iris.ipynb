```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

***1. Read iris dataset into the program***

```python
df = pd.read_excel("C:\\Users\\Sopanam\\Downloads\\Assignments\\iris.xls")
```

***2. Display dataset's columns***

```python
print(df.columns)
```

***3. Display entire dataset***

```python
print(df)
```

***4. Mean of each column***

```python
mean_sl = df['SL'].mean()
print("mean of SL=",mean_sl)
mean_sw = df['SW'].mean()
print("mean of SW =",mean_sw)
mean_pl = df['PL'].mean()
print("mean of PL=",mean_pl)
mean_pw = df['PW'].mean()
print("mean of PW=",mean_pw)
```

***5. Calculate the mean of all Numeric Columns***

```python
df.mean(numeric_only=True)
```

***6. Check for null values present in the dataset***

```python
df.isnull().sum()
```

## Plots ##

***1. Scatter Plot between sw and sl of different species of iris***

```python
sns.scatterplot(x='SL',y='SW',hue='Classification',data=df)
plt.show()
```

***2. Plot between PL and PW for different iris***

```python
sns.scatterplot(x='PL',y='PW',hue='Classification',data=df)
plt.show()
```

***3. Histogram of SL of different species of iris***

```python
sns.histplot(x='SL',hue='Classification',data=df)
plt.show()
```

***4. Relational Plot***

```python
sns.relplot(data=df, x="SL", y="Classification", hue="Classification")
```

```python
sns.pairplot(df, hue='Classification')
plt.show()
```
